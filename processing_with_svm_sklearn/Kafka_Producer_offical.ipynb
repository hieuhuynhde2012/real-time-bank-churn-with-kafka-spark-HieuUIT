```python
import json
import time
import pandas as pd
from kafka import KafkaProducer

# Cu hnh Kafka Producer
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)

file_path = 'C:/Users/PC/Desktop/Do_an_Big_data/PythonCodes/work/data/Churn_Modelling_Sampled.csv'
data = pd.read_csv(file_path)

try:
    while True:
        row = data.sample(n=1).iloc[0]  # Chn ngu nhin mt dng

        # Chuyn i d liu nhng gi nguyn kiu int, float, bool
        transaction = {
            col: (None if pd.isna(row[col]) else
                  bool(row[col]) if data[col].dtype == 'bool' else
                  int(row[col]) if data[col].dtype == 'int64' else
                  float(row[col]) if data[col].dtype == 'float64' else
                  str(row[col]))
            for col in data.columns
        }

        # Kim tra kiu d liu trc khi gi
        # print("\n--- Kim tra kiu d liu trc khi gi ---")
        # for key, val in transaction.items():
        #     print(f"{key}: {val} ({type(val).__name__})")

        # Gi d liu n Kafka
        try:
            producer.send('producer_svm', value=transaction)
            print(f"Sent: {transaction}")
        except Exception as e:
            print(f"Li khi gi d liu: {e}")

        time.sleep(2)  # Gi lp gi d liu theo thi gian thc

except KeyboardInterrupt:
    print("\nDng gi d liu Kafka.")
finally:
    producer.close()
```
