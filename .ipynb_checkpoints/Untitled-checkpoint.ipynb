```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
import pygad
import pygad.nn
import pygad.gann
import numpy as np

# Khởi tạo Spark Session
spark = SparkSession.builder \
    .appName("GeneticNN_Training") \
    .getOrCreate()
```

```python
# Đọc dữ liệu tĩnh để đào tạo mô hình GNN
df = spark.read \
    .format("csv") \
    .option("header", "true") \
    .schema("CustomerId STRING, CreditScore INT, Geography STRING, Gender STRING, Age INT, Balance DOUBLE, NumOfProducts INT, HasCrCard INT, IsActiveMember INT, EstimatedSalary DOUBLE, Exited INT") \
    .load("Churn_Modelling.csv")
```

```python
# Tiền xử lý dữ liệu
from pyspark.ml.feature import StringIndexer, VectorAssembler

indexers = [
    StringIndexer(inputCol=col_name, outputCol=col_name + "_index")
    for col_name in ["Geography", "Gender"]
]
assembler = VectorAssembler(
    inputCols=["CreditScore", "Geography_index", "Gender_index", "Age", "Balance", "NumOfProducts", "HasCrCard", "IsActiveMember", "EstimatedSalary"],
    outputCol="features"
)
```

```python
from pyspark.ml import Pipeline
pipeline = Pipeline(stages=indexers + [assembler])
processed_df = pipeline.fit(df).transform(df)

# Chuyển dữ liệu về dạng numpy để dùng với pygad
features = np.array(processed_df.select("features").collect())
labels = np.array(processed_df.select("Exited").collect())
```
